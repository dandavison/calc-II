```python
%matplotlib inline
%config InlineBackend.figure_formats = {'svg',}
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate data to plot
x = np.linspace(0, 2 * np.pi, 200)
y = np.exp(np.sin(x))

# Make plot
plt.plot(x, y)
plt.xlim((0, 2 * np.pi))
plt.xlabel(r'$x$')
plt.ylabel(r'$\mathrm{e}^{\sin{x}}$')
```
