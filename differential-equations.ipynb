```python
%matplotlib inline
%config InlineBackend.figure_formats = {'svg',}

from functools import partial

import matplotlib.pyplot as plt
import numpy as np
```

```python
# 9.1.6
def fn(x, C):
    return (np.log(x) + C) / x


x = np.linspace(0.1, 2, 100)
plt.plot(x, fn(x, 0), 'b',
         x, fn(x, 1), 'g',
         x, fn(x, 2), 'r')
```

```python
# 9.1.8
def fn(x, c):
    return (c - x**2)**(-1/2)

x = np.linspace(-2, 2, 100)
plt.plot(x, fn(x, 0.1), 'b',
         x, fn(x, 1), 'g',
         x, fn(x, 2), 'r')
```
