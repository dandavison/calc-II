```python
%matplotlib inline
%config InlineBackend.figure_formats = {'svg',}

import matplotlib.pyplot as plt
import numpy as np

from matplotlib_utils import plot_direction_field
```

```python
# 9.1.6
def fn(x, C):
    return (np.log(x) + C) / x


x = np.linspace(0.1, 2, 100)
plt.plot(x, fn(x, 0), 'b',
         x, fn(x, 1), 'g',
         x, fn(x, 2), 'r')
```

```python
# 9.1.8
def fn(x, c):
    return (c - x**2)**(-1/2)

x = np.linspace(-2, 2, 100)
plt.plot(x, fn(x, 0.1), 'b',
         x, fn(x, 1), 'g',
         x, fn(x, 2), 'r',
         x, fn(x, 1/4), 'k')
plt.axhline(y=2, color='k', ls='--')
```

```python
# 9.3.2

def deriv1(x, y):
    return x * np.exp(-y)

def soln(x, C):
    return np.log(x**2/2 + C)

plot_direction_field(deriv1, (-3, 3), (-3, 3))
xs = np.linspace(-3, 3, 61)
for C in np.linspace(-5, 4, 9):
    plt.plot(xs, soln(xs, C))
```

```python
# 9.3.4

def deriv1(x, y):
    return 1/(1 + x) * 1/(y ** 2)

def soln(x, C):
    return np.power(3 * np.log(1 + x) + C, 1/3)

plot_direction_field(deriv1, (-3, 3), (-3, 3))
xs = np.linspace(-3, 3, 61)
for C in np.linspace(-5, 5, 10):
    plt.plot(xs, soln(xs, C))
```

```python
# 9.3.6

def deriv1(s, v):
    return (1 + s) / (s*(1 + v))

def soln(s, C, sign):
    return -1 + sign * np.sqrt(2*s + 2*np.log(s) + C)

plot_direction_field(deriv1, (-3, 3), (-3, 3))
ss = np.linspace(-3, 3, 60)
for C in np.linspace(-5, 5, 10):
    plt.plot(ss, soln(ss, C, 1))
```

```python
# 9.3.10

def deriv1(t, z):
    return -np.exp(t + z)

def soln(t, C):
    return -np.log(np.exp(t) + C)

plot_direction_field(deriv1, (-3, 3), (-3, 3))
ss = np.linspace(-3, 3, 60)
for C in np.linspace(-5, 5, 10):
    plt.plot(ss, soln(ss, C))
```

```python
# 9.3.12

def deriv1(x, y):
    return np.log(x) / (x * y)

def soln(x, C):
    return np.sqrt(np.log(x)**2 + 4)

plot_direction_field(deriv1, (-3, 3), (-3, 3))
ss = np.linspace(-3, 3, 60)
for C in np.linspace(-5, 5, 10):
    plt.plot(ss, soln(ss, C))
    
plt.plot([1], [2], "o")
```

```python
# 9.3.12

def deriv1(x, y):
    return -y/x

def orth(x, C):
    return np.sqrt(x**2 + C)

plot_direction_field(deriv1, (-3, 3), (-3, 3))
ss = np.linspace(-3, 3, 60)
for C in np.linspace(-5, 5, 10):
    plt.plot(ss, orth(ss, C))
    plt.plot(ss, -orth(ss, C))
```
